Import all our libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

load in csv data

```python
df = pd.read_csv('well_data.csv')
```

```python
X = df['DEPT'].values
y = df['HCAL'].values
plt.scatter(X,y, alpha=0.035, color='blue')
plt.xlabel('Well Depth')
plt.ylabel('HCAL')
```

From Here I predict that there will be a correlation, however it may not be as strong

```python
from sklearn.linear_model import LinearRegression
```

```python
X = X.reshape(-1,1)
regression = LinearRegression()
regression.fit(X, y)
```

```python
regression.score(X, y) # Retruns 0.32 for R^2
```

```python
plt.scatter(X, y, alpha=0.035, c='blue')
plt.plot(X, regression.predict(X), c='red', linewidth=2.50)
```
